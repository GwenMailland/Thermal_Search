```python
import gpxpy

def lire_fichier_gpx(chemin_fichier):
    try:
        # Lecture du fichier GPX
        with open(chemin_fichier, 'r') as gpx_file:
            gpx = gpxpy.parse(gpx_file)
        
        # Initialisation des listes pour stocker les donnes
        points = []
        times = []
        longitude = []
        latitude = []
        elevations = []
        speeds = []
        speeds_elevation = []
        
        
        # Extraction des donnes des points
        for track in gpx.tracks:
            for segment in track.segments:
                previous_point = None  # Point prcdent pour le calcul de la vitesse
                previous_elevations = None
                for point in segment.points:
                    points.append(point)
                    times.append(point.time)
                    latitude.append(point.latitude)
                    longitude.append( point.longitude)
                    elevations.append(point.elevation)
                    # Calcul de la vitesse entre deux points
                    speed = None
                    if previous_point is not None:
                        speed = point.speed_between(previous_point)
                    speeds.append(speed)
                    previous_point = point

                    speed_elevation = None
                    if previous_elevations is not None:
                        speed_elevation = point.elevation-previous_elevations
                    speeds_elevation.append(speed_elevation)
                    previous_elevations = point.elevation

        
        # Cration du tableau de donnes
        data = {
            'time': times,
            'latitude': latitude,
            'longitude': longitude,
            'altitude': elevations,
            'speed': speeds,
            'speed_elevation': speeds_elevation
        }
        df = pd.DataFrame(data)
        return df
    except FileNotFoundError:
        print("Le fichier spcifi n'a pas t trouv.")
        return None
    except Exception as e:
        print("Une erreur s'est produite lors de la lecture du fichier:", str(e))
        return None

# Exemple d'utilisation
chemin_fichier_gpx = '/Users/gwen/Desktop/Thermal_Search/Data_St-Hilaire>20km_100p/AirDhuile+2426189/2426189.gpx'
donnees_gpx = lire_fichier_gpx(chemin_fichier_gpx)

donnees_gpx
donnees_gpx.to_csv('data.csv', index=False)
```
